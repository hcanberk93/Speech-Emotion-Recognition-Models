```python
def getListOfFiles(dirName,orgClasses = False):
    # create a list of file and sub directories 
    # names in the given directory 
    listOfFile = sorted(os.listdir(dirName))
    allFiles = list()
    # Iterate over all the entries
    for entry in listOfFile:
        # Create full path
        fullPath = os.path.join(dirName, entry)
        # If entry is a directory then get the list of files in this directory 
        if os.path.isdir(fullPath):
            allFiles = allFiles + getListOfFiles(fullPath,orgClasses)
        else:
            if orgClasses == True:
                fileName = entry.split('/')[-1]; #Choosing the filename only
                fileName = entry.split('.')[0].split('-'); #Removing .wav then splitting info numbers
                if int(fileName[2])>1 and int(fileName[2])<7:
                    allFiles.append(fullPath)
            else:
                allFiles.append(fullPath)                
    return allFiles
```

```python
def convertToDigital(fileList):
    Xlist = np.zeros((len(fileList),int(sampleRate*2.5)))
    index = 0
    for X in fileList:
        N, sample_rate = librosa.load(X, res_type='kaiser_fast',duration=2.5,sr=sampleRate,offset=0.5)
        Xlist[index,0:len(N)]=N
        index = index + 1
    return Xlist
```

```python
def extractFeatures(fileList,col):
    featureList = np.zeros((len(fileList),col))
    ind = 0
    for entry in fileList:
        wavOrig,_ = librosa.load(entry, res_type='kaiser_fast',duration=2.5,sr=sampleRate,offset=0.5)
        spectogram = np.abs(librosa.stft(wavOrig)**2)
        #spectogram = spectogram.clip(min=0)
        mel = librosa.filters.mel(sr=sampleRate,n_fft=2048, n_mels=40)       
        melSpectogram = mel.dot(spectogram)
        melSpectogram = librosa.feature.spectral_rolloff(S=melSpectogram, sr=sampleRate, roll_percent=0.95)
        melSpectogram = librosa.power_to_db(melSpectogram,ref=np.median)
        #M *= (1+ (lifter / 2)* np.sin(np.pi * np.arange(1, 1 + n_mfcc, dtype=M.dtype) / lifter)[:, np.newaxis])
        
        mfcc = dct(melSpectogram, type=2, axis=0, norm='ortho')
        featureList[ind,0:mfcc.shape[1]] = mfcc
    return featureList
```

```python
def extractMFCC(digitalWavs):
    mfccFeatures = np.zeros((np.size(digitalWavs,0),216))
    index = 0
    for X in digitalWavs:
        mfccFeatures[index,:] = np.mean(librosa.feature.mfcc(y=X, sr=sampleRate, n_mfcc=40, hop_length=512),axis=0)
        index = index +1
    return mfccFeatures
```

```python
def classifyData(fileName,cFunc=True,OrgResearchClasses = False): #boolList:boolen val that defines if used for singular or list form
    #3 Emotion (1-Neutral 2-Calm 3-Happy 4-Sad 5-Angry 6-Scared 7-Disgust 8-Surprised)
    #4 Emotional Intensity(1-Normal 2-Strong)
    #7 Actor (%2 == 1 Male %2 == 0 Female)
        ravFeature = list();
        fileList = getListOfFiles(fileName)
        if OrgResearchClasses == False:
            for entry in fileList:
                ravFeature.append(extractFromFName(entry,cFunc));
            return ravFeature;
        else:
            for entry in fileList:
                orgClasses = {'Female_Calm':1,'Male_Calm':2,'Female_Happy':3,'Male_Happy':4,'Female_Sad':5,
                              'Male_Sad':6,'Female_Angry':7,'Male_Angry':8,'Female_Fearful':9,'Male_Fearful':10};
                labels = {1:"Neutral", 2:"Calm", 3:"Happy", 4:"Sad", 5:"Angry", 6:"Fearful", 7:"Disgust", 8:"Surprised"};
                entry = entry.split('/')[-1]; #Choosing the filename only
                entry = entry.split('.')[0].split('-'); #Removing .wav then splitting info numbers
                if int(entry[2])>1 and int(entry[2])<7 :
                    genderPrefix = "Male" if int(entry[6])%2 ==1 else "Female"
                    emotionPrefix = labels[int(entry[2])];
                    className = genderPrefix + '_'+emotionPrefix
                    classIndex = orgClasses[className]
                    ravFeature.append(classIndex)
            
            return ravFeature

def extractFromFName(fileName,cFunc):
    #cFunc: Boolean val that enables/disables the classifyFunc
    fileName = fileName.split('/')[-1]; #Choosing the filename only
    fileName = fileName.split('.')[0].split('-'); #Removing .wav then splitting info numbers
    fileName = [int(i) for i in fileName]; #Converting string array into int array
    if(cFunc == True):
        val = classifyFunc([fileName[2],fileName[3],fileName[6]%2]);
    else:
        val = [fileName[2],fileName[3],fileName[6]%2];
    return val
    
def classifyFunc(nameArr):
    emClass = ['Neutral','Calm','Happy','Sad','Angry','Fearful','Disgust','Surprised'];
    em,emInt,gender = emClass[nameArr[0]-1],('Normal', 'Strong') [nameArr[1]%2==0],\
    ('Male', 'Female') [nameArr[2]%2==0];
    nameArr = [em,emInt,gender];
    return nameArr;
```

```python
def dataNormalization(X):
    mean = np.mean(X, axis=0);
    std = np.std(X,axis=0);
    X = (X-mean)/std;
    return X;
```

```python
class cnn:
    def __init__(self, labels, features, test_size, num_classes=8, epochs=500,
                 batch_size=96):  # not sure if ideal // hyperparmeter tuning needed
        self.labels = labels
        self.features = features
        self.test_size = test_size
        self.x_train = None
        self.y_train = None
        self.x_test = None
        self.y_test = None
        self.model = None
        self.num_classes = num_classes
        self.epochs = epochs
        self.batch_size = batch_size

        
        
    def split_train_test(self):
            self.x_train, self.x_test, self.y_train, self.y_test = sklearn.model_selection.train_test_split(
                self.features, self.labels, test_size=self.test_size, random_state=np.random)
    
    def build_model(self):
        layers = keras.layers
        self.y_train = keras.utils.to_categorical(self.y_train, self.num_classes)
        self.y_test = keras.utils.to_categorical(self.y_test, self.num_classes) 
    
        self.model = keras.Sequential([
            layers.Conv1D(256,5,padding='same',input_shape=self.x_train.shape[1:],activation=tf.nn.relu),
            layers.Conv1D(128,5,padding='same',activation=tf.nn.relu),
            layers.Dropout(0.1),
            layers.MaxPooling1D(pool_size=(8)),
            layers.Conv1D(128,5,padding='same',activation=tf.nn.relu),
            layers.Conv1D(128,5,padding='same',activation=tf.nn.relu),
            layers.Conv1D(128,5,padding='same',activation=tf.nn.relu),
            layers.Dropout(0.2),
            layers.Conv1D(128,5,padding='same',activation=tf.nn.relu),
            layers.Flatten(),
            layers.Dense(self.num_classes, activation='softmax'),
        ])
        # initiate RMSprop optimizer
        opt = keras.optimizers.RMSprop(lr=0.00001, decay=1e-6)
        # Let's train the model using RMSprop
        self.model.compile(loss='categorical_crossentropy',
                           optimizer=opt,
                           metrics=['accuracy'])
        print("built cnn model..")
        print(self.model.summary())

    def train(self):       
        print("training model")
        # may want to change this
        save_dir = os.path.join(os.getcwd(), 'Gumelar_Saved_Models')
        model_name = 'keras_cnn_trained_model.h5'
        # model_name = 'cnn_trained_model-{epoch:02d}-{val_acc:.2f}.h5'
        if not os.path.isdir(save_dir):
            os.makedirs(save_dir)
        model_path = os.path.join(save_dir, model_name)
        # if os.path.isfile(model_path):
        #     self.model.load_weights(model_path)
        # may want to change save_best to true
        checkpoint = ModelCheckpoint(
            model_path, monitor='val_acc', verbose=1, save_best_only=False, mode='max')
        tensorboard = TensorBoard(log_dir="logs/{}".format(time()), write_graph=True,
                                  write_images=True, histogram_freq=0)
        
        history = self.model.fit(self.x_train, self.y_train,
                    batch_size=self.batch_size,
                    epochs=self.epochs,
                    validation_data=(self.x_test, self.y_test),
                    shuffle=True,
                    callbacks=[tensorboard, checkpoint])
        
        return history

    def printHistory(self,history):
        import matplotlib.pyplot as plt
        from matplotlib.pyplot import figure

        figure(figsize=(8, 6), dpi=80)
        acc = history.history['accuracy']
        val_acc = history.history['val_accuracy']
        loss = history.history['loss']
        val_loss = history.history['val_loss']
        
        epochs = range(1, len(acc) + 1)
        plt.plot(epochs, loss, 'r-', label='Eitim Kayb')
        plt.plot(epochs, val_loss, 'b-', label='Dorulama Kayb')
        plt.title('Eitim ve Dorulama Kayb')
        plt.xlabel('Epok')
        plt.ylabel('Kayp')
        plt.legend()
        plt.show()

        figure(figsize=(8, 6), dpi=80)
        plt.plot(epochs, acc, 'r-', label='Eitim Baarm')
        plt.plot(epochs, val_acc, 'b-', label='Dorulama Baarm')
        plt.title('Eitim ve Dorulama Baarm')
        plt.xlabel('Epok')
        plt.ylabel('Baarm')
        plt.legend()
        plt.show()
        
    def evaluate(self):
        
        scores = self.model.evaluate(self.x_test, self.y_test, verbose=1)
        predictions = self.model.predict_classes(self.x_test)
        
            
        print('Test loss:', scores[0])
        print('Test accuracy:', scores[1])

        from sklearn.metrics import classification_report
        
        # add roc_auc
        y_new = np.argmax(self.y_test, axis=1)
        print(np.shape(predictions), np.shape(y_new))
        report = classification_report(y_new, predictions)
        print(report)

        
        matrix = confusion_matrix(y_new, predictions)
        print (matrix)
        df_cm = pd.DataFrame(matrix, range(self.num_classes),
                  range(self.num_classes))
        # plt.figure(figsize = (10,7))
        sn.set(font_scale=1.4)#for label size
        sn.heatmap(df_cm, annot=True,annot_kws={"size": 12})# font size
        plt.show()
        plt.savefig('confusion_matrix.pdf')
        plt.savefig('confusion_matrix.pgf') # for latex
```
